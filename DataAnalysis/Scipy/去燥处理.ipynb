```python
# coding:utf-8
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from scipy import fftpack as fft
```

```python
path = r'E:\python\PythonSpace\Git\Python\DataAnalysis\Data\dog.jpg'
dog = plt.imread(path)
plt.imshow(dog)
plt.show(dog)
```

```python
dog.shape
```

```python
# Scipy
dog_fft = fft.fft2(dog)
dog_fft
```

```python
dog_fft.max()
```

```python
dog_fft.min()
```

```python
# 
dog_fft2 = np.where(np.abs(dog_fft)>3e6, 0, dog_fft)
```

```python
# 
dog_ifft = fft.ifft2(dog_fft2)
# dog_ifft
```

```python
# 
dog_real = np.real(dog_ifft)
# dog_real
```

```python
plt.imshow(dog_real)
plt.show(dog_real)
```
