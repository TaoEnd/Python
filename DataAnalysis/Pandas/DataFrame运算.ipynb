```python
# coding:utf-8
import numpy as np
import pandas as pd
from pandas import DataFrame
```

```python
df1 = DataFrame(np.random.randint(0, 150, size=(4, 4)), 
				index=["张三", "李四", "王五", "赵六"],
				columns=["语文", "数学", "英语", "Python"])
df1
```

```python
df2 = DataFrame(np.random.randint(0, 150, size=(5, 4)),
				index=["张三", "李四", "王五", "赵六", "Mike"],
				columns=["语文", "数学", "英语", "Python"])
df2
```

```python
# 没有同时出现在df1和df2中的，就默认为NaN
df1 + df2
```

```python
# 行索引不对应时，默认添上NaN
df3 = df1.add(df2, fill_value=0)
df3
```

```python
# 列索引不对应时，也默认添上NaN
```

```python
s1 = df3["语文"]
s1
```

```python
s2 = df3.loc["张三"]
s2
```

```python
# 列索引相同时，二维的DataFrame可以和一维的Series相加
# 此时使用的广播模式，DataFrame的每一行都加上了Series
df3 + s2
```

```python
# axis=0（或"index"）表示按列进行相加，也使用了广播机制，DataFrame的每一列都加上了Series
# axis=1（或"columns"）表示按行进行相加
df3.add(s1, axis=0)
```
