```python
# pandas处理缺失值
df = DataFrame({"age" : [20, 18, 26, 22, 24],
				"salary" : [10000, 12000, 9700, 11000, 8900]},
				index=["张三", "李四", "王五", "赵六", "Mike"],
				columns=["age", "salary", "work"])
df
```

```python
df.work["李四" : "赵六"] = "Python"
df
```

```python
# any(axis=0)与any()相同，表示对于某一列而言，只要有元素为True,
# 那这一列就为True
# any(axis=1)表示对于某一行而言，只要有元素为True，那这一行就为True
s1 = df.isnull().any(axis=1)
df[s1]
```

```python
# all(axis=0)与all()相同，表示对于某一列而言，必须所有的元素都为True,
# 这一列元素才为True
# all(axis=1)表示对于某一行而言，只有所有元素为True，这一行才为True
s2 = df.notnull().all(axis=1)
df[s2]
```

```python
# 过滤函数
# dropna()默认axis=0，表示当某一行中有NaN类型数据时，就删除这一行
# axis=1时，表示当某一列中有NaN类型数据时，就删除这一列
df.dropna()
```

```python
# how="all"表示只有某行或者某列的全部数据都为NaN类型时，才删除
# how默认是"any"
df.dropna(how="all")
```

```python
# 数据填充，对所有空数据统一替换
# 将NaN类型数据填充成java
df.fillna(value="Java")
```

```python
# 向前填充(method="ffill")和向后填充(method=backfill)
# 即使用前一条非空数据填充当前数据，还是使用后一条非空数据填充当前数据
df.fillna(method="backfill")
```

```python
df.fillna(method="ffill")
```

```python
# 填充时，默认复制一份原数据，不对原数据进行修改
# inplace=True时，就会修改原数据
df.fillna(method="ffill", inplace=True)
df
```
