```python
# coding:utf-8
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
df1 = DataFrame({"employee" : ["Tom", "Jack", "Lucy"],
				 "group" : ["sail", "counting", "market"]})
df2 = DataFrame({"employee" : ["Tom", "Jack", "Lucy"],
				 "work_time" : [1.5, 2, 1]})
df1
```

```python
df2
```

```python
# （一对一）合并操作，按照相同列进行合并，不相同的则直接去除
pd.merge(df1, df2)
```

```python
df2 = DataFrame({"employee" : ["Tom", "Tom", "Lucy"],
				 "work_time" : [1.5, 2, 1]})
df2
```

```python
pd.merge(df1, df2)
```

```python
df1 = DataFrame({"employee" : ["Tom", "Jack", "Lucy"],
				 "group" : ["sail", "counting", "market"]})
df1
```

```python
pd.merge(df1, df2)
```

```python
# 当df1和df2中有两列相同，可以使用Key指定按照哪一列进行合并
df3 = DataFrame({"employee" : ["Tom", "Jack", "Lucy"],
				 "group" : ["sail", "counting", "market"],
				 "salary" : [1200, 1000, 1300]})
df3
```

```python
df4 = DataFrame({"employee" : ["Tom", "Mike", "Lucy"],
				 "team" : ["market", "it", "market"],
				 "work_time" : [2, 3, 1.5]})
df4
```

```python
# on指定了按照哪一列数据进行合并
# suffixes表示为冲突的列，添加后缀重新命名
pd.merge(df3, df4, on = "employee", suffixes = ["A", "B"])
```

```python
# 在实际中，两个DataFrame的列名可能不一样，但是表示的
# 意思是一样的，在合并时，可以使用left_on和right_on
# 进行指明，设置名字不同但是意义相同的列
pd.merge(df3, df4, left_on="group", right_on="team")
```

```python
# 内合并和外合并，内合并只保留相同的列，是merge()的默认
# 模式，外合并会保留不同的列，使用NaN进行填充
df1 = DataFrame({"age": [18, 22, 26], "height": [177, 168, 173]})
df1
```

```python
df2 = DataFrame({"age": [18, 24, 25], "weight": [68, 57, 73]})
df2
```

```python
# 默认内合并
pd.merge(df1, df2)
```

```python
# 外合并，没有的元素使用NaN填充
pd.merge(df1, df2, how = "outer")
```

```python
# left join
pd.merge(df1, df2, how = "left")
```
