```python
# coding:utf-8
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
nd = np.random.randint(0, 10, size=(3, 3))
nd
```

```python
np.concatenate((nd, nd), axis=1)
```

```python
df1 = DataFrame(np.random.randint(0, 150, size=(4, 4)),
				index = list("ABCD"), columns = list("一二三四"))
df1
```

```python
df2 = DataFrame(np.random.randint(0, 150, size=(4, 4)),
				index = list("efgh"), columns = list("一二三四"))
df2
```

```python
# 默认axis=0, 从行上进行级联
# 某个DataFrame中没有出现元素时，则默认填上NaN
pd.concat([df1, df2],axis=1)
```

```python
df3 = DataFrame(np.random.randint(0, 150, size=(4, 4)),
				index = list("ABCD"), columns = list("一二三四"))
df3
```

```python
pd.concat([df1, df3], axis=1)
```

```python
# 在级联时，忽略原索引，并重新设置，
# 重新设置时，两个DataFrame合并在一起，然后从0开始排序
pd.concat([df1, df3], axis=1, ignore_index=True)
```

```python
# 在级联时，使用keys可以增加多级索引
pd.concat([df1, df3], keys=["张三", "李四"])
```

```python
df4 = DataFrame(np.random.randint(0, 150, size=(3, 4)),
				index = list("ABC"), columns = list("1234"))
df4
```

```python
# 外链接，级联时，默认是外链接
# join="inner"表示内连接，只连接相同索引的项
pd.concat([df1, df4], axis=1, join='inner')
```

```python
# 左外连接（只能在行上进行连接），以左方DataFrame的列为基础进行连接
pd.concat([df1, df4], join_axes=[df1.columns], join='inner')
```

```python
# 使用append()追加数据
df1.append(df2)
```
