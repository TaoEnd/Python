```python
# coding:utf-8
from matplotlib import pyplot as plt
# 使用魔法指令
%matplotlib inline
```

```python
path = r'E:\python\PythonSpace\Git\Python\DataAnalysis\Data\Numpy\14.jpg'
image = plt.imread(path)
plt.imshow(image)
# 这一句代码有错，但是不写这一句又无法展示图片
plt.show(image)
```

```python
# 将整个图像进行上下颠倒
image1 = image[::-1]
plt.imshow(image1)
plt.show(image1)
```

```python
# 将整个图片即上下颠倒，也左右颠倒
image2 = image[::-1, ::-1]
plt.imshow(image2)
plt.show(image2)
```

```python
# 将整个图片的颜色进行颠倒
image3 = image[::, ::, ::-1]
plt.imshow(image3)
plt.show(image3)
```

```python
# 隔50步取一个数据
image4 = image[::50, ::50]
plt.imshow(image4)
plt.show(image4)
```

```python
# 从第100行开始，每隔5步就设置为0
image[100::5,::5] = 0
plt.imshow(image)
plt.show(image)
```
