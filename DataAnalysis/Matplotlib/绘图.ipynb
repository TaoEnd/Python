```python
# coding:utf-8
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(0, 100, 100)
plt.figure(figsize=(5, 5))
plt.plot(x, x*x)
plt.show()
```

```python
x = np.arange(0, 10, 1)
plt.plot(x, x*x)
plt.plot(x, x*x/2)
plt.show()
```

```python
# 
x = np.arange(-2*np.pi, 2*np.pi, np.pi/100)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.grid(True)
plt.show()
```

```python
# 
plt.figure(figsize=(15, 3))

fig1 = plt.subplot(1, 3, 1)
x = np.arange(-2*np.pi, 2*np.pi, np.pi/100)
fig1.grid(color="r", linestyle="--", linewidth=2)
fig1.plot(x, np.sin(x))

fig2 = plt.subplot(1, 3, 3)
x = np.arange(-2*np.pi, 2*np.pi, np.pi/100)
fig2.grid(linestyle="-.", linewidth=2)
fig2.plot(x, np.cos(x))

plt.show()
```

```python
# 
plt.figure(figsize=(15, 3))

fig1 = plt.subplot(1, 3, 1)
x = np.arange(-2*np.pi, 2*np.pi, np.pi/100)
# xy
fig1.axis([-6, 6, -1.1, 1.1])
fig1.grid(color="r", linestyle="--", linewidth=2)
fig1.plot(x, np.sin(x))

fig2 = plt.subplot(1, 3, 3)
x = np.arange(-2*np.pi, 2*np.pi, np.pi/100)
fig2.grid(linestyle="-.", linewidth=2)
fig2.plot(x, np.cos(x))

plt.show()
```

```python
# 
x = np.linspace(0, 20, 40)
plt.axis("off")
plt.plot(x)
plt.show()
```

```python
x = np.linspace(-4, 4, 100)
plt.axis("equal")
plt.plot(np.sin(x), np.cos(x))
plt.show()
```

```python
x = np.arange(-2*np.pi, 2*np.pi, np.pi/100)
plt.title("f(x) = sin(x)", fontsize=20)
plt.grid(color="r", linestyle="--", linewidth=2)
plt.xlim(-6, 6)
plt.ylim(-1.1, 1.1)
plt.xlabel("x label", fontsize=15)
plt.ylabel("y label", fontsize=15)
plt.plot(x, np.sin(x))
plt.show()
```

```python
# 
x = np.arange(-2*np.pi, 2*np.pi, np.pi/100)
plt.title("f(x)", fontsize=20)
plt.xlim(-6, 6)
plt.ylim(-1.1, 1.1)
plt.xlabel("x label", fontsize=15)
plt.ylabel("y label", fontsize=15)

# loc=

# line1, = plt.plot(x, np.sin(x))
# line2, = plt.plot(x, np.cos(x))
# plt.legend([line1, line2], ["sin", "cos"], loc="best")

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.legend(["sin", "cos"], loc="best")

plt.show()
```

```python
# ncol
x = np.arange(-2*np.pi, 2*np.pi, np.pi/100)
plt.title("f(x)", fontsize=20)
plt.xlim(-6, 6)
plt.ylim(-1.4, 1.1)
plt.xlabel("x label", fontsize=15)
plt.ylabel("y label", fontsize=15)

# loc=

# line1, = plt.plot(x, np.sin(x))
# line2, = plt.plot(x, np.cos(x))
# plt.legend([line1, line2], ["sin", "cos"], loc="best")

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.legend(["sin", "cos"], loc="best", ncol=2)

plt.show()
```

```python
# linestlemarker
# marker
x1 = np.random.randn(50)
x2 = np.random.randn(50)
x3 = np.random.randn(50)
# cumsum()
plt.plot(x1.cumsum(), color="red", linestyle="--", marker="o")
plt.plot(x2.cumsum(), color="green", linestyle=":", marker="+")
plt.plot(x3.cumsum(), color="blue", linestyle="-.")
plt.legend(["x1", "x2", "x3"])
plt.show()
```
