```python
# coding:utf-8
import scipy.misc as misc
import numpy as np
from matplotlib import pyplot as plt
```

```python
# 图片的灰度处理，将图片展示成黑白的
face = misc.face(gray=True)
plt.imshow(face, cmap="gray")
plt.show(face)
```

```python
path = r'E:\python\PythonSpace\Git\Python\DataAnalysis\Data\dog.jpg'
dog = plt.imread(path)
plt.imshow(dog)
plt.show(dog)
```

```python
# 图片从三维降为二维的，则可以从彩色的变成黑白的
# 使用最大值法（或者最小值法)
dog_data = dog.max(axis=2)
plt.imshow(dog_data, cmap="gray")
plt.show(dog_data)
```

```python
# 使用平均值法
dog_data2 = dog.mean(axis=2)
plt.imshow(dog_data2, cmap="gray")
plt.show(dog_data2)
```

```python
# 加权平均
w = np.array([0.299, 0.587, 0.114])
dog_data3 = np.dot(dog, w)
plt.imshow(dog_data3, cmap="gray")
plt.show(dog_data3)
```
