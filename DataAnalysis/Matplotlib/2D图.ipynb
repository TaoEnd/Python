```python
# coding:utf-8
import numpy as np
from matplotlib import pyplot as plt
```

```python
# 直方图
x = np.random.randint(0, 10, 10)
plt.figure(figsize=(10, 5))
# 使用rwidth调整直方图中每条线的宽度
plt.hist(x, rwidth=0.5)
plt.show()
```

```python
# 条形图：bar、barh
x = np.linspace(0, 5, 5)
y = np.random.randint(0, 20, size=5)
plt.bar(x, y)
plt.show()
```

```python
# 饼图
p = np.array([70, 33, 20, 10])
plt.figure(figsize=(4, 4))
# "%1.2f%%"：2表示保留小数位数，%%表示以百分号形式展示
plt.pie(p, labels=["dog", "cat", "duck", "other"], autopct="%1.2f%%")
plt.show()
```

```python
# 展示不完整的饼图，没有达到1或者100时，缺少的直接显示空白
p = np.array([0.7, 0.2, 0.05, 0.04])
plt.figure(figsize=(4, 4))
plt.pie(p, labels=["dog", "cat", "duck", "other"], autopct="%1.2f%%")
plt.show()
```

```python
p = np.array([0.4, 0.3, 0.15, 0.15])
plt.figure(figsize=(4, 4))
plt.pie(p, labels=["dog", "cat", "duck", "other"], 
		labeldistance=1.05, autopct="%1.2f%%")
plt.show()
```

```python
p = np.array([0.4, 0.3, 0.15, 0.15])
plt.figure(figsize=(4, 4))
plt.pie(p, labels=["dog", "cat", "duck", "other"], 
		explode=[0.1, 0, 0, 0], autopct="%1.2f%%")
plt.show()
```

```python
p = np.array([0.4, 0.3, 0.15, 0.15])
plt.figure(figsize=(4, 4))
plt.pie(p, labels=["dog", "cat", "duck", "other"], 
		explode=[0.1, 0, 0, 0], shadow=True, autopct="%1.2f%%")
plt.show()
```

```python
# 散点图
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.scatter(x, y, s = 5)
plt.show()
```

```python
# 图片注解
x = np.arange(-np.pi, np.pi, np.pi/100)
plt.plot(x, np.sin(x))
plt.text(-3, 0, "sin(x)")
plt.show()
```

```python
# 添加注释
x = np.random.randint(0, 10, 10)
plt.plot(x)
# 添加注释
# xytext设置文本位置
# arrowprops设置箭头类型
plt.annotate(s="this point is important", xy=(4, x[4]), 
			 xytext=(5, x[4]+1), 
			 arrowprops={"width": 1, "headwidth": 10, "headlength": 10})
plt.show()
```
